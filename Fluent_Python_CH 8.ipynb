**Example 8-5 :  The Relative Immutability of Tuples **

- Immutability of a tuple referes to the physical contents of tuple data structure and not the referenced objects inside.
- Identity of items in a tuple can never change.
- 

```
t1 = (1, 2, [30, 40])
t2 = (1, 2, [30, 40])
```

```
t1 == t2
```

```
t1 is t2
```

```
id(t1[-1])
```

```
t1[-1].append(99)
t1
```

```
id(t1[-1])
```

```
t1 == t2
```

**Conclusion :** Now that a mutable item (i.e. a list) has changed inside t1 tuple, t1 and t2 are different!
